```python
import numpy as np 
import pandas as pd 
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
data = pd.read_csv("../input/fetal-health-classification/fetal_health.csv")
data.head()
```

```python
data.info()
```

```python
from sklearn.utils import resample
df_1=data[data["fetal_health"]==1]
df_2=data[data["fetal_health"]==2]
df_3=data[data["fetal_health"]==3]

df_1_upsample=resample(df_1,replace=True,n_samples=2000,random_state=123)
df_2_upsample=resample(df_2,replace=True,n_samples=2000,random_state=124)
df_3_upsample=resample(df_3,replace=True,n_samples=2000,random_state=125)

data=pd.concat([df_1_upsample,df_2_upsample,df_3_upsample])
```

```python
data['fetal_health'].value_counts()
```

```python
X=data.drop(["fetal_health"],axis=1)
y=data["fetal_health"]

#Set up a standard scaler for the features
col_names = list(X.columns)
s_scaler = preprocessing.StandardScaler()
X_df= s_scaler.fit_transform(X)
X_df = pd.DataFrame(X_df, columns=col_names)   
X_df.describe().T
```

```python
#spliting test and training sets
X_train, X_test, y_train,y_test = train_test_split(X_df,y,test_size=0.3333,random_state=42)
```

```python
Xtrain= X_train.values.reshape((len(X_train),21,1))
Xtrain.shape
```

```python
Xtest= X_test.values.reshape((len(X_test),21,1))
Xtest.shape
```

```python
from keras.utils import to_categorical
import numpy
ytrain = to_categorical(y_train.values)
ytrain.shape
```

```python
ytest= y_test.values.reshape((len(y_test.values),1))
ytest = to_categorical(ytest)
ytest.shape
```

```python
from keras.models import Sequential
from keras.optimizers import SGD,Adam
from keras.layers import LSTM, Dense, Flatten, Activation, MaxPool1D, Dropout
from keras.layers.normalization import BatchNormalization
from keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
models = Sequential()
models.add(LSTM(units=128, return_sequences=True, input_shape=(21,1), activation='relu'))
models.add(BatchNormalization())
models.add(MaxPool1D(pool_size=(3), strides=(2), padding="same"))
models.add(LSTM(units=64, return_sequences=True, input_shape=(21,1), activation='relu'))
models.add(BatchNormalization())
models.add(MaxPool1D(pool_size=(2), strides=(2), padding="same"))
models.add(Flatten())
models.add(Dense(32, activation='relu'))
models.add(Dense(4, activation='softmax'))


models.compile(loss='categorical_crossentropy',
              optimizer='adam',
               metrics=['acc'])

callbacks = [EarlyStopping(monitor='val_loss', patience=8), ModelCheckpoint(filepath='CTG_resampled.h5', monitor='val_loss', save_best_only=True)]
history=models.fit(Xtrain, ytrain, epochs=20, batch_size=32,callbacks=callbacks, validation_data=(Xtest,ytest))
```

```python
from sklearn.metrics import classification_report

predz = models.predict(Xtest).argmax(axis=-1)
print(classification_report(y_test,predz) )
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,predz)
```
